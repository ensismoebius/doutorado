```python
import os
import matplotlib.pyplot as plt
import numpy as np

%matplotlib widget
```

```python
def curve(value):
    return value ** 2
```

```python
def curveDerivative(value):
    return 2 * value
```

```python
# 100 linearly spaced numbers
xAxis = np.linspace(-np.pi,np.pi,100)

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Plot the set of the tangent lines in all defined points
# if y=mx+n then:
for p in xAxis:
    m = curveDerivative(p) # m is angular coefficient of the equation (derivative)
    n = -curve(p) # n is the linear coefficient given by the value of the function in the give point
    plt.plot(xAxis, m*xAxis+n, 'c') # plotting the resulting tangents

# plot the functions
plt.plot(xAxis,curve(xAxis), 'b', label='Curve')

plt.legend(loc='upper left')

# show the plot
plt.show()
```
